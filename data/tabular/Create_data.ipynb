```python
import random
import math
import seaborn as sbs
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(21) ## must be together with the generator
n = 10000
x = np.random.uniform(0,1, n) # x coordinate 
y = np.random.uniform(0,1, n) # y coordinate 
y_sin = np.sin(x * np.pi) # sin value for x, which is used to determine type 

initial_positive = y < y_sin
concept_drift_positive = y > y_sin
```

```python
# Create a DataFrame for Seaborn plotting
data = {
    'x': x,
    'y': y,
    'Classification': np.where(initial_positive, 'Initially Positive', 'Initially Negative'),
    'Concept Drift': np.where(concept_drift_positive, 'After Concept Drift Positive', 'After Concept Drift Negative')
}
df = pd.DataFrame(data)

# Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='x', y='y', hue='Classification', data=df, palette=['blue', 'red'])
sns.scatterplot(x='x', y='y', hue='Concept Drift', data=df, palette=['green', 'orange'], marker='x')
sns.lineplot(x=x, y=y_sin, color='black', label='y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Classification of Points Relative to y = sin(x)')
plt.legend()
plt.show()
```

```python
df.to_csv('Sine.csv')
```

## Mixed

```python
np.random.seed(21) ## must be together with the generator
n = 10000
x = np.random.uniform(0,1, n) # x coordinate 
y = np.random.uniform(0,1, n) # y coordinate 
y_sin = np.sin(x * 3 * np.pi) # sin value for x, which is used to determine type 

initial_positive = y < 0.5 + 0.3 * y_sin
concept_drift_positive = y > 0.5 + 0.3 * y_sin
```

```python
data = {
    'x': x,
    'y': y,
    'Classification': np.where(initial_positive, 'Initially Positive', 'Initially Negative'),
    'Concept Drift': np.where(concept_drift_positive, 'After Concept Drift Positive', 'After Concept Drift Negative')
}
df = pd.DataFrame(data)
```

```python
df.to_csv('Mixed.csv')
```
